### 01. Hodrick Prescott Filter: Separate Trend & Cycle


<div class="alert alert-info"><h3>Related Function:</h3>
<tt><a href='https://www.statsmodels.org/stable/generated/statsmodels.tsa.filters.hp_filter.hpfilter.html'><strong>statsmodels.tsa.filters.hp_filter.hpfilter</strong></a><font color=black>(X, lamb=1600)</font>&nbsp;&nbsp;Hodrick-Prescott filter</div>
    
The <a href='https://en.wikipedia.org/wiki/Hodrick%E2%80%93Prescott_filter'>Hodrick-Prescott filter</a> separates a time-series  $y_t$ into a trend component $\tau_t$ and a cyclical component $c_t$

$y_t = \tau_t + c_t$

The components are determined by minimizing the following quadratic loss function, where $\lambda$ is a smoothing parameter:

$\min_{\\{ \tau_{t}\\} }\sum_{t=1}^{T}c_{t}^{2}+\lambda\sum_{t=1}^{T}\left[\left(\tau_{t}-\tau_{t-1}\right)-\left(\tau_{t-1}-\tau_{t-2}\right)\right]^{2}$


The $\lambda$ value above handles variations in the growth rate of the trend component.<br>When analyzing quarterly data, the default lambda value of 1600 is recommended. Use 6.25 for annual data, and 129,600 for monthly data.

```python
import numpy as py
import pandas as pd
%matplotlib inline

#for import data direct from API
#import statsmodels.api as sm
#df = sm.datasets.macrodata.load_pandas().data
#df.index = pd.Index(sm.tsa.datetools.dates_from_range('1959Q1', '2009Q3'))
#print(sm.datasets.macrodata.NOTE)

df = pd.read_csv('../Data/macrodata.csv',index_col=0,parse_dates=True)
df.head()
```

```python
#For using HP filter: to get trend and cycle
from statsmodels.tsa.filters.hp_filter import hpfilter

# Tuple unpacking
gdp_cycle, gdp_trend = hpfilter(df['realgdp'], lamb=1600)
```

```python
#for drop kolom ---> df.drop(columns=['trend'])
selected_columns = df[["realgdp","year","quarter"]]
df_gdp = selected_columns.copy()
#df_gdp.head()

df_gdp['trend'] = gdp_trend
df_gdp['cycle'] = gdp_cycle

df_gdp.head()
```

```python
#for spesific index
df_gdp[['realgdp','trend']]['2001-01-01':].plot(figsize=(15,4)).autoscale(axis='x',tight=True)
#df[['trend','realgdp']].plot().autoscale(axis='x',tight=True);
```

```python
#for cycle
df_gdp['cycle']['2001-01-01':].plot(figsize=(15,4)).autoscale(axis='x',tight=True)
```

### 02. ETS Decomposition: Error, Trend & Seasonality


Statsmodels provides a <em>seasonal decomposition</em> tool we can use to separate out the different components. This lets us see quickly and visually what each component contributes to the overall behavior.


* We apply an <strong>additive</strong> model when it seems that the trend is more linear and the seasonality and trend components seem to be constant over time (e.g. every year we add 10,000 passengers).<br>
* A <strong>multiplicative</strong> model is more appropriate when we are increasing (or decreasing) at a non-linear rate (e.g. each year we double the amount of passengers).

Keep it mind, in a time-series, consist of: trend, cycle, seasonality also error (residual)

what we got from this method:
* observer = the real data
* trend = trend of data (T) / a whole trend. Usually a linear
* seasonality = cycle data of spesific time (S) (usuallay < 1 year)
* residual = error data, or residual data from trend & seasonality

```python
#try to use another dataset, it was a airline dataset
df_airline = pd.read_csv('Data/airline_passengers.csv',index_col='Month',parse_dates=True)
df_airline.dropna()
df_airline.head()
```

```python
df_airline.plot();
```

```python
#import ets decomposition
from statsmodels.tsa.seasonal import seasonal_decompose

#for lock the plot
from pylab import rcParams
rcParams['figure.figsize'] = (15,4)

df_gdp_result = seasonal_decompose(df_gdp['realgdp'], model="add")  # model='mul' add mul
df_gdp_result.plot();
```

```python
df_airline_result = seasonal_decompose(df_airline['Thousands of Passengers'], model="mul")  # model='mul' also works
df_airline_result.plot();
```

### 03. EWMA: Exponential-Weighted Moving Average


Shopisticated model of SMA (simple moving average), that has weakness such as:
* the entire model will be constrained by same window size (60, 120 etc)
* that we know, smaller windows size will lead to more noise (not a signal)
* always lag by window size
* it will never reach to full the peak or valley of data due it only averaging
* also it doesnt really inform possible future data behavior, only describe data's trend
* sometimes, extreme historical data (like recession) would be make a skew data significantly

EWMA will reduce the effect of those issues by put more weight value (applying more weight to more recent values). And the amount of weight depend on the actual parameters and number of periods (window size)

```python
#import the matplotlib
import matplotlib.pyplot as plt

df_airline.head(5)
```

```python
#try a simple SMA
df_airline_sma_6 = df_airline['6-month-SMA'] = df_airline['Thousands of Passengers'].rolling(window=6).mean()
df_airline_sma_12 = df_airline['12-month-SMA'] = df_airline['Thousands of Passengers'].rolling(window=12).mean()
#df_airline_sma_24 = df_airline['24-month-SMA'] = df_airline['Thousands of Passengers'].rolling(window=24).mean()

df_airline['6-month-SMA'] = df_airline_sma_6
df_airline['12-month-SMA'] = df_airline_sma_12

#for deleting columns
#df_airline.drop(['24-month-SMA'], axis=1, inplace=True)
df_airline.head()
```
