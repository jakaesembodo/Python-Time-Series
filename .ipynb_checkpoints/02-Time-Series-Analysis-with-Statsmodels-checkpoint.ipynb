### 01. Hodrick Prescott Filter: Separate Trend & Cycle


<div class="alert alert-info"><h3>Related Function:</h3>
<tt><a href='https://www.statsmodels.org/stable/generated/statsmodels.tsa.filters.hp_filter.hpfilter.html'><strong>statsmodels.tsa.filters.hp_filter.hpfilter</strong></a><font color=black>(X, lamb=1600)</font>&nbsp;&nbsp;Hodrick-Prescott filter</div>
    
The <a href='https://en.wikipedia.org/wiki/Hodrick%E2%80%93Prescott_filter'>Hodrick-Prescott filter</a> separates a time-series  $y_t$ into a trend component $\tau_t$ and a cyclical component $c_t$

$y_t = \tau_t + c_t$

The components are determined by minimizing the following quadratic loss function, where $\lambda$ is a smoothing parameter:

$\min_{\\{ \tau_{t}\\} }\sum_{t=1}^{T}c_{t}^{2}+\lambda\sum_{t=1}^{T}\left[\left(\tau_{t}-\tau_{t-1}\right)-\left(\tau_{t-1}-\tau_{t-2}\right)\right]^{2}$


The $\lambda$ value above handles variations in the growth rate of the trend component.<br>When analyzing quarterly data, the default lambda value of 1600 is recommended. Use 6.25 for annual data, and 129,600 for monthly data.

### 02. ETS Decomposition: Error, Trend & Seasonality


Statsmodels provides a <em>seasonal decomposition</em> tool we can use to separate out the different components. This lets us see quickly and visually what each component contributes to the overall behavior.


* We apply an <strong>additive</strong> model when it seems that the trend is more linear and the seasonality and trend components seem to be constant over time (e.g. every year we add 10,000 passengers).<br>
* A <strong>multiplicative</strong> model is more appropriate when we are increasing (or decreasing) at a non-linear rate (e.g. each year we double the amount of passengers).

Keep it mind, in a time-series, consist of: trend, cycle, seasonality also error (residual)

what we got from this method:
* observer = the real data
* trend = trend of data (T) / a whole trend. Usually a linear
* seasonality = cycle data of spesific time (S) (usuallay < 1 year)
* residual = error data, or residual data from trend & seasonality

### 03. EWMA: Exponential-Weighted Moving Average


Shopisticated model of SMA (simple moving average), that has weakness such as:
* the entire model will be constrained by same window size (60, 120 etc)
* that we know, smaller windows size will lead to more noise (not a signal)
* always lag by window size
* it will never reach to full the peak or valley of data due it only averaging
* also it doesnt really inform possible future data behavior, only describe data's trend
* sometimes, extreme historical data (like recession) would be make a skew data significantly

EWMA will reduce the effect of those issues by put more weight value (applying more weight to more recent values). And the amount of weight depend on the actual parameters and number of periods (window size)

Heres the formula for ALPHA

\begin{split}\alpha =
 \begin{cases}
     \frac{2}{s + 1},               & \text{for span}\ s \geq 1\\
     \frac{1}{1 + c},               & \text{for center of mass}\ c \geq 0\\
     1 - \exp^{\frac{\log 0.5}{h}}, & \text{for half-life}\ h > 0
 \end{cases}\end{split}

* <strong>Span</strong> corresponds to what is commonly called an N-day EW moving average. -> number of days
* <strong>Center of mass</strong> has a more physical interpretation and can be thought of in terms of span: $c=(s1)/2$
* <strong>Half-life</strong> is the period of time for the exponential weight to reduce to one half.
* <strong>Alpha</strong> specifies the smoothing factor directly. from 0 to 1

We have to pass precisely one of the above into the <tt>.ewm()</tt> function. For our data we'll use <tt>span=12</tt>.

```python
#for exponential-weight moving average (EWMA)
df_airline['EWMA12'] = df_airline['Thousands of Passengers'].ewm(span=12,adjust=False).mean()
df_airline['EWMA12'].head()
```

```python
#comparng EWMA and SMA
df_airline[['Thousands of Passengers','EWMA12','12-month-SMA']].plot(figsize=(12,8)).autoscale(axis='x',tight=True);
```

### 04. HOLT-WINTERS Method


In the previous section on <strong>Exponentially Weighted Moving Averages</strong> (EWMA) we applied <em>Simple Exponential Smoothing</em> using just one smoothing factor $\alpha$ (alpha). This failed to account for other contributing factors like trend and seasonality.

In this section we'll look at <em>Double</em> and <em>Triple Exponential Smoothing</em> with the <a href='https://otexts.com/fpp2/holt-winters.html'>Holt-Winters Methods</a>. 

In <strong>Double Exponential Smoothing</strong> (aka Holt's Method) we introduce a new smoothing factor $\beta$ (beta) that addresses trend:

\begin{split}l_t &= (1 - \alpha) l_{t-1} + \alpha x_t, & \text{    level}\\
b_t &= (1-\beta)b_{t-1} + \beta(l_t-l_{t-1}) & \text{    trend}\\
y_t &= l_t + b_t & \text{    fitted model}\\
\hat y_{t+h} &= l_t + hb_t & \text{    forecasting model (} h = \text{# periods into the future)}\end{split}

Because we haven't yet considered seasonal fluctuations, the forecasting model is simply a straight sloped line extending from the most recent data point. We'll see an example of this in upcoming lectures.

With <strong>Triple Exponential Smoothing</strong> (aka the Holt-Winters Method) we introduce a smoothing factor $\gamma$ (gamma) that addresses seasonality:

\begin{split}l_t &= (1 - \alpha) l_{t-1} + \alpha x_t, & \text{    level}\\
b_t &= (1-\beta)b_{t-1} + \beta(l_t-l_{t-1}) & \text{    trend}\\
c_t &= (1-\gamma)c_{t-L} + \gamma(x_t-l_{t-1}-b_{t-1}) & \text{    seasonal}\\
y_t &= (l_t + b_t) c_t & \text{    fitted model}\\
\hat y_{t+m} &= (l_t + mb_t)c_{t-L+1+(m-1)modL} & \text{    forecasting model (} m = \text{# periods into the future)}\end{split}

Here $L$ represents the number of divisions per cycle. In our case looking at monthly data that displays a repeating pattern each year, we would use $L=12$.

In general, higher values for $\alpha$, $\beta$ and $\gamma$ (values closer to 1), place more emphasis on recent data.

```python
# SIMPLE EXPONENCIAL SMOOTING = EWMA

from statsmodels.tsa.holtwinters import SimpleExpSmoothing

span = 12
alpha = 2/(span+1) #alpha formula using span in EWM

#df_airline['EWMA12'] = df_airline['Thousands of Passengers'].ewm(alpha=alpha,adjust=False).mean()
df_airline['SES12']=SimpleExpSmoothing(df_airline['Thousands of Passengers']).fit(smoothing_level=alpha,optimized=False).fittedvalues.shift(-1)
df_airline.head()
```

```python
df_airline[['Thousands of Passengers','EWMA12','SES12']].iloc[:24].plot(figsize=(12,8)).autoscale(axis='x',tight=True);
```

from the above, we learn that EWM is similar with Holt Winter with simple exponential smoothing (only use alpha).

##### Double Exponential Smoothing

Where Simple Exponential Smoothing employs just one smoothing factor    (alpha), Double Exponential Smoothing adds a second smoothing factor    (beta) that addresses trends in the data. Like the alpha factor, values for the beta factor fall between zero and one ( 0<1 ). The benefit here is that the model can anticipate future increases or decreases where the level model would only work from recent calculations.

We can also address different types of change (growth/decay) in the trend. If a time series displays a straight-line sloped trend, you would use an additive adjustment. If the time series displays an exponential (curved) trend, you would use a multiplicative adjustment.

As we move toward forecasting, it's worth noting that both additive and multiplicative adjustments may become exaggerated over time, and require damping that reduces the size of the trend over future periods until it reaches a flat line.

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

df_airline['DES_ADD_12'] = ExponentialSmoothing(df_airline['Thousands of Passengers'], trend='add').fit().fittedvalues.shift(-1)
df_airline.head()
```

```python
df_airline[['Thousands of Passengers','SES12','DES_ADD_12']].iloc[:12].plot(figsize=(15,4)).autoscale(axis='x',tight=True)
```

```python
df_airline['DES_MUL_12'] = ExponentialSmoothing(df_airline['Thousands of Passengers'], trend='mul').fit().fittedvalues.shift(-1)
df_airline.head()
```

```python
df_airline[['Thousands of Passengers','SES12','DES_ADD_12','DES_MUL_12']].iloc[:12].plot(figsize=(15,4)).autoscale(axis='x',tight=True)
```

from the above, the precision model is using multiple adjustment better than additive adjustment.

##### Triple Exponential Smoothing
the method most closely associated with Holt-Winters, adds support for both trend (level and trend) and seasonality in the data. 


```python
df_airline['TES_ADD_12'] = ExponentialSmoothing(df_airline['Thousands of Passengers'],trend='add',seasonal='add',seasonal_periods=12).fit().fittedvalues
df_airline['TES_MUL_12'] = ExponentialSmoothing(df_airline['Thousands of Passengers'],trend='mul',seasonal='mul',seasonal_periods=12).fit().fittedvalues
df_airline.head()
```

```python
df_airline[['Thousands of Passengers','SES12','DES_ADD_12','DES_MUL_12','TES_ADD_12','TES_MUL_12']].iloc[:12].plot(figsize=(15,4)).autoscale(axis='x',tight=True)
```

```python
df_airline.head()
```

```python
#measure the differences

df_airline['%_6-month-SMA'] = (df_airline['6-month-SMA']-df_airline['Thousands of Passengers']).abs()/df_airline['Thousands of Passengers']
df_airline['%_12-month-SMA'] = (df_airline['12-month-SMA']-df_airline['Thousands of Passengers']).abs()/df_airline['Thousands of Passengers']
df_airline['%_24-month-SMA'] = (df_airline['24-month-SMA']-df_airline['Thousands of Passengers']).abs()/df_airline['Thousands of Passengers']
df_airline['%_EWMA12'] = (df_airline['EWMA12']-df_airline['Thousands of Passengers']).abs()/df_airline['Thousands of Passengers']
df_airline['%_SES12'] = (df_airline['SES12']-df_airline['Thousands of Passengers']).abs()/df_airline['Thousands of Passengers']
df_airline['%_DES_ADD_12'] = (df_airline['DES_ADD_12']-df_airline['Thousands of Passengers']).abs()/df_airline['Thousands of Passengers']
df_airline['%_DES_MUL_12'] = (df_airline['DES_MUL_12']-df_airline['Thousands of Passengers']).abs()/df_airline['Thousands of Passengers']
df_airline['%_TES_ADD_12'] = (df_airline['TES_ADD_12']-df_airline['Thousands of Passengers']).abs()/df_airline['Thousands of Passengers']
df_airline['%_TES_MUL_12'] = (df_airline['TES_MUL_12']-df_airline['Thousands of Passengers']).abs()/df_airline['Thousands of Passengers']
df_airline.head(10)
```

```python
print('SMA 6 month ==',df_airline['%_6-month-SMA'].mean())
print('SMA 12 month ==',df_airline['%_12-month-SMA'].mean())
print('SMA 24 month ==',df_airline['%_24-month-SMA'].mean())
print('EWM 12 month ==',df_airline['%_EWMA12'].mean())
print('SES 12 month ==',df_airline['%_SES12'].mean())
print('DES Additive 12 month ==',df_airline['%_DES_ADD_12'].mean())
print('DES Multiplicative 12 month ==',df_airline['%_DES_MUL_12'].mean())
print('TES Additive 12 month ==',df_airline['%_TES_ADD_12'].mean())
print('TES Multiplicative 12 month ==',df_airline['%_TES_MUL_12'].mean())
```
